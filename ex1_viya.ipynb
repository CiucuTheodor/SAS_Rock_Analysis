```sas
#R
# Install haven package (if not installed)
if (!require("haven")) install.packages("haven")

library(haven)

# Read the SAS file
df <- read_sas("data/courses/ACADEMIC/hotel_booking_demand_dataset.sas7bdat")

# Display the first 5 rows
head(df, 5)
# Statistics
summary(df)
```

```sas
*SAS;

libname mydata '/export/viya/homes/anda.belciu@ie.ase.ro/Courses/ACADEMIC';
title 'First 5 obs';
proc print data=mydata.hotel_booking_demand_dataset (obs=5);
run;

proc means data=mydata.hotel_booking_demand_dataset n mean std min max;
title "Statistics - numerical var";
run;

proc freq data=mydata.hotel_booking_demand_dataset;
tables _character_ / nocum nopercent;
title "Frequencies for categorical var";
run;
```
